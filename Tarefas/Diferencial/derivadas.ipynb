# Eros Moreira Ferreira

```python
import numpy as np
import matplotlib.pyplot as plt

def Dif_frente(f, x, h):
    """retorna a derivada de f usando aproximação x+h """
    return (f(x + h) - f(x)) / h

def Dif_atras(f, x, h):
    """retorna a derivada de f usando aproximação x-h """
    return (f(x) - f(x - h)) / h 

def Dif_centrada(f, x, h):
    """retorna a derivada de f usando aproximação x-h e x+h

    Args:
        f (função): função a ser derivada
        x (float): ponto em que a derivada deve ser calculada 
    """
    return (f(x + h) - f(x - h)) / (2 * h)



```

```python
senx = lambda x: np.sin(x)
```

```python

plt.figure(figsize = (15, 20),dpi = 300)
plt.figtext(0.5, 0.01, "Diferentes aproximações númericas da derivada de $sen(x)$ " , position= (0.1,0.9), fontsize=20)
plot1 = plt.subplot2grid((2, 2), (0, 0), xlabel = 'x', ylabel = 'y') 
plot2 = plt.subplot2grid((2, 2), (1, 0), xlabel = 'x', ylabel = 'y', colspan=2) 
plot3 = plt.subplot2grid((2, 2), (0, 1), xlabel = 'x', ylabel = 'y') 

# Calculo dos pontos
for N in range(3,11): # Valores do N
    h = (2*np.pi) / 2**N
    x = np.arange(0, 2*np.pi+h, h)
    y_centrada = [Dif_centrada(senx, xi, h) for xi in x]
    y_frente = [Dif_frente(senx, xi, h) for xi in x]
    y_atras = [Dif_atras(senx, xi, h) for xi in x]
    
    
    plot1.plot(x,y_frente, linewidth = 0.5, label = f"{N = }")
    plot2.plot(x,y_centrada, linewidth = 0.5, label = f"{N = }")
    plot3.plot(x,y_atras, linewidth = 0.5,label = f"{N = }")

# Config do gráfico
xtikcs = ["0", r"$\frac{\pi}{3}$", r"$\frac{2\pi}{3}$", r"$\pi$",r"$\frac{4\pi}{3}$",r"$\frac{5\pi}{3}$", r"$2\pi$"]

plot1.plot(x,np.cos(x),'--', linewidth = 0.5, label = "cos(x)")
plot1.set_xticks(np.linspace(0,2*np.pi,7), xtikcs)
plot1.set_title("Diferencial frente")
plot1.legend()

plot2.plot(x,np.cos(x),'--', linewidth = 0.5, label = "cos(x)")
plot2.set_xticks(np.linspace(0,2*np.pi,7), xtikcs)
plot2.set_title("Diferencial centrada")
plot2.legend()

plot3.plot(x,np.cos(x),'--', linewidth = 0.5, label = "cos(x)")
plot3.set_xticks(np.linspace(0,2*np.pi,7), xtikcs)
plot3.set_title("Diferencial Atrás")
plot3.legend()


plot1.axhline(color = 'gray', zorder = -4, linewidth = 1)
plot2.axhline(color = 'gray', zorder = -4, linewidth = 1)
plot3.axhline(color = 'gray', zorder = -4, linewidth = 1)

plot1.axvline(color = 'gray', zorder = -4, linewidth = 1)
plot2.axvline(color = 'gray', zorder = -4, linewidth = 1)
plot3.axvline(color = 'gray', zorder = -4, linewidth = 1)
# plt.savefig('derivadas.pdf')
plt.show()
```

```python
x = np.pi
h =[ (2*np.pi) / 2**N for N in range(3,11)]
erro_centrada = [abs(Dif_centrada(senx, x, h) - np.cos(x)) for h in h]
erro_frente = [abs(Dif_frente(senx, x, h) - np.cos(x)) for h in h]
erro_atras = [abs(Dif_atras(senx, x, h) - np.cos(x)) for h in h]

h_log = np.log2(np.array(h))

plt.xlabel(r'$\log_2h$')
plt.ylabel('erro')
plt.plot(h_log,erro_centrada, label = 'centrada')
plt.plot(h_log,erro_frente, '--',label = 'frente')
plt.plot(h_log,erro_atras,'-.', linewidth = 0.5,label = 'atras')
plt.legend()

# Só exibe uma reta quando os dois eixos estão em log
# --------------------------
# plt.xlabel(r'$\log_2h$')
# plt.ylabel('$log_2erro$')
# plt.loglog(h,erro_centrada, label = 'centrada')
# plt.loglog(h,erro_frente, '--',label = 'frente')
# plt.loglog(h,erro_atras,'-.', linewidth = 0.5,label = 'atras')
# plt.legend()
# --------------------------



plt.show()
```
